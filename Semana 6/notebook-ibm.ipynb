Ahora tratemos de agrupar datos por medio de diferentes categorías. Para el próximo ejemplo, buscaremos los valores únicos para una de las variables del conjunto de datos.

```python
df['drive-wheels'].unique()
```

Seleccionemos las variables que queremos utilizar

```python
df_group_one = df[['drive-wheels','body-style','price']]
df_group_one.head()
```

Ahora agrupamos por drive-wheels usando el método groupby

```python
df_group_one = df_group_one.groupby(['drive-wheels'],as_index=False).mean()
df_group_one
```

También se puede agrupar por múltiples variables

```python
df_gptest = df[['drive-wheels','body-style','price']]
grouped_test1 = df_gptest.groupby(['drive-wheels','body-style'],as_index=False).mean()
grouped_test1
```

también podemos pivotear la tabla, es decir, pasar las filas a columnas y las columnas a filas.

```python
grouped_pivot = grouped_test1.pivot(index='drive-wheels',columns='body-style')
grouped_pivot
```

Por último, podemos visualizar la relación de cada categoría por medio del siguiente gráfico.

```python
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot, cmap='RdBu')

#Nombre de los labels
row_labels = grouped_pivot.columns.levels[1]
col_labels = grouped_pivot.index

#Mover las marcas y labels al centro
ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5, minor=False)

#insertar labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotar los labels si son muy largos
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```

Del gráfico anterior se observa que hatchback y 4wd corresponde al menor precio.

**Anova**

El ANOVA es un método estadístico para probar si existen diferencias significativas en las medias de dos o más grupos. El anova devuelve dos parámetros:

* P-Value: Determina que la significancia estadística de la prueba realizada
* Prueba F: Es una métrica, donde mientras más grande es la métrica, mayor es la diferencia entre las medias.
